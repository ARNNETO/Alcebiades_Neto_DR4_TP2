* Tom: Usuário pergunta de forma clara e amigável

```python
# Importa bibliotecas
from dotenv import load_dotenv
import os
import google.generativeai as genai
import os

# Lê o arquivo .env
load_dotenv()
```

```python
# Notícia que será avaliada
noticia = """
A Xiaomi lançou nesta quarta-feira o celular Redmi K80 Pro em edição inspirada na Lamborghini. O dispositivo representa a continuidade da parceria com a fabricante de carros esportivos, apresentando um visual ousado com cores vibrantes.
Assim como já havia sido observado na edição especial do K70 Pro, o novo modelo Lamborghini SQUADRA CORSE contará com uma versão em verde, de tom bastante chamativo. No entanto, nesta edição, a variante amarela foi substituída pela preta, que confere um toque mais clássico ao produto.
Além disso, o acabamento dos smartphones destaca a divisão em formato de Y, agora mais visível nos novos modelos e com detalhes na cor vermelha. A logomarca da Lamborghini está posicionada na parte inferior, identificando os dispositivos.
Suas configurações são semelhantes às do K80 Pro tradicional, também apresentado nesta quarta-feira. Ele será vendido exclusivamente na versão com as especificações máximas do celular, incluindo 16 GB de RAM e 1 TB de armazenamento interno.
O desempenho do Redmi K80 Pro é liderado por um processador Snapdragon 8 Elite, a plataforma mobile mais potente da Qualcomm no momento. Implementada também no Xiaomi 15, ela oferece até 45% mais performance em comparação com a geração anterior.
Outras características do aparelho incluem:
Bateria de 6.000 mAh com suporte para recargas de 120 W com cabo e 50 W sem fio
Câmera principal de 50 MP, ultrawide de 32 MP e teleobjetiva 2,5x de 50 MP
Câmera frontal de 20 MP
Tela TCL M9 OLED com resolução de 3.200 x 1.440 pixels, suporte para taxas de atualização de 120 Hz e brilho máximo de 3.200 nits
Sistema operacional Android 15, com HyperOS 2.0
Preço e disponibilidade
O “celular da Lamborghini” da Xiaomi foi apresentado ao mercado chinês, com preço oficial de R$ 4.999 yuan (cerca de R$ 4.014 em conversão direta). Não há previsão para vendas em outros países, embora o K80 Pro padrão deva ser renomeado com a marca da POCO para comercialização global.
"""

# Cria o prompt 
prompt = f"""
Faça um resumo da <NOTICIA> seguindo os seguintes exemplos:

<NOTICIA>
"A Xiami lançou no dia dd/mm/aaaa o aparelho X"
###
<RESUMO>
"A empresa Xiami apresentou seu mais novo smartphone, o X."

Agora, faça o resumo da noticia abaixo:
<NOTICIA>
"{noticia}"
###
<RESUMO>
"""

# Executao prompt usando o  modelo Gemini
genai.configure(api_key=os.environ["GEMINI_KEY"])
model = genai.GenerativeModel("gemini-1.5-flash")
response = model.generate_content(prompt)

# Apresenta a resposta
# print("O resumo foi:")
# print(response)

# Extrai apenas o texto
result = response.candidates[0].content.parts[0].text

# Exibe resultado
print("O resumo da notícia foi:\n")
print(result)
```
